### Importing Libraries

```python
import pandas as pd
import numpy as np 
import warnings 
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='ISO-8859-1')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# number of missing values in each column of our dataframe
dataset.isnull().sum()
```

```python
# total number of cells having missing values in our dataset
missing_values = (dataset.isnull().sum()).sum()
missing_values
```

```python
# total number of cells in our dataframe 
total_cells = dataset.size
total_cells
```

```python
# percentage of missing values in our dataframe
(missing_values/total_cells)*100
```

Since more than half of the values in our dataframe are missing so we would be using selective columns only.
First we will rename the columns which we would be using to make it readable and more interpretable.

```python
dataset.rename(columns = {'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                          'region_txt':'Region','attacktype1_txt':'Attack_Type','target1':'Target','nkill':'Killed',
                          'nwound':'Wounded','gname':'Group','targtype1_txt':'Target_Type',
                          'weaptype1_txt':'Weapon_type','latitude':'Latitude',
                          'longitude':'Longitude','target1':'Target','city':'City'}, inplace=True)
```

Creating new dataframe with necessary columns which we would be using.

```python
# creating dataframe with necessary columns only
df = dataset[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','Attack_Type','Killed',
              'Wounded','Group','Target','Target_Type','Weapon_type']]
df.head()
```

```python
df.info()
```

```python
df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df['Killed'] = df['Killed'].fillna(0).astype(int)
```

```python
df.info()
```

### Cleaned Data

```python
df.head()
```

```python
df.nunique()
```

## Data Visualization

### Most attacked target types

```python
# 10 most attacked targets

plt.figure(figsize = (11,6))
sns.barplot(df['Target_Type'].value_counts().head(10).index, df['Target_Type'].value_counts().head(10).values, 
            palette='magma')
plt.title('Top 10 most attacked targets',fontsize=15)
plt.xlabel('Targets',fontsize=15)
plt.ylabel('Number of attacks',fontsize=15)
plt.xticks(rotation=70)
plt.show()
```

### Regions attacked most

```python
# regions with most attacks 

plt.figure(figsize=(13,7))
sns.barplot(df['Region'].value_counts().index, df['Region'].value_counts().values,palette='flare')
plt.title('Most Attacked Regions',fontsize=15)
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation=70)
plt.show()
```

Middle East and North African Region is most affected.

### Years in which Middle East and North African region was attacked most

```python
# creating dataframe for Middle East and North Africa 
mena = df[df.Region == 'Middle East & North Africa']
```

```python
# attacks over the years in MENA 
plt.figure(figsize=(15,7))
sns.barplot(mena['Year'].value_counts().index, mena['Year'].value_counts().values)
plt.title('Terror Attacks over the years in Middle East & North African Region',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

The highest number of attacks in the MENA region were during 2014 and after that numbers are decreasing.

### Countries Attacked most

```python
# top 10 countries attacked most
plt.figure(figsize=(13,7))
sns.barplot(df['Country'].value_counts().head(10).index, df['Country'].value_counts().head(10).values)
plt.title('Top 10 Most Attacked Countries',fontsize=15)
plt.xlabel('Countries',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
```

### Number of attacks in Iraq over the years

```python
# creating a separate dataframe for Iraq
df_Iraq = df[df['Country'] == 'Iraq']
```

```python
# Number of attacks in Iraq over the years 
plt.figure(figsize = (15,7))
sns.barplot(df_Iraq['Year'].value_counts().index, df_Iraq['Year'].value_counts().values,palette = 'deep')
plt.title('Terror Attacks in Iraq over the Years',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

In the year 2003 terror attacks in Iraq started gaining traction. It was steadily increasing till year 2012 and the next 2 years there was a drastic increase, maxing out during year 2014. After 2014 a downward trend can be seen but in 2017 the number attacks was still very high which was around 2500 which means that on average there were 7 attacks per day.

### Conclusion

After performing the Exploratory Data Analysis we get the following insights from the data:
1. Private Citizens and Property were attacked most followed by Military, Police, Government and so on.
2. Middle East & North Africa was most affected among the top affected region. Most of the people in this region were either wounded or killed.
3. Iraq was the country which was most affected by terror attacks and had maximum number of killed and wounded people.
4. The State and City that was most affected was Baghdad
5. In the last decade 2014 had most number of terror attacks. That was around 16500 attacks during this year which means on average 45 attacks per day.
6. The most common attack type was Bombing/Explosion.
7. Explosives have been consistently the most popular weapon of choice for terrorists.
8. Taliban became more active since 2012 and they are responsible for the most of the terror attacks

### Recommendation

1. Since Private Citizens and Property are being targeted consistently so stronger security and surveillance should be provided, especially in the densely populated regions.
2. More surveillance is required especially in the Middle East & North African Regions.
3. Strict border policy should be implemented to prevent the movement of explosives between the regions.
